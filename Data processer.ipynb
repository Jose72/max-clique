```python
import csv
with open('results.txt', 'r') as results_file:
    results_iter = csv.DictReader(results_file, delimiter=',', quotechar='|')
    results = list(results_iter)
```

```python
import matplotlib.pyplot as plt
nodes = list(map(lambda x: int(x['graph_nodes']), results))
edges = list(map(lambda x: int(x['graph_edges']), results))
hits_triangles = list(map(lambda x: int(x['hits_triangle']), results))
count_of_cliques_received = list(map(lambda x: int(x['count_of_cliques_received']), results))
calls_made = list(map(lambda x: int(x['calls_made']), results))
densities = list(map(lambda x: x[0]/x[1] if x[1] != 0 else -1, zip(nodes, edges)))
```

```python
plt.plot(densities, calls_made, 'o')
plt.xscale('log')
plt.yscale('log')
plt.show()
```
